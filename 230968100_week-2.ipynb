### 2) For each of the numeric attributes (proteins up to vitamins), identify and replace all missing data (indicated with -1) with the arithmetic mean of the attribute.

```python
data = data.replace("?",-1)
data = data.mask(data == -1)
for i in data.columns:
    if data[i].isnull().any():
        data[i] = data[i].fillna(data[i].astype(float).mean())
data
```

### 4) For each of the numeric attributes (proteins up to vitamins), identify and replace all noisy data with the median of attribute. 

```python
num=data.select_dtypes(include=['int64', 'float64']).columns
for column in num:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median = data[column].median()
    data[column] = data[column].apply(lambda x: median if x < lower_bound or x > upper_bound else x)
data
```

### Compute the percentage of missing values in the DataFrame. If a column has more than 15% missing values then drop (or remove) the column from the DataFrame.

```python
df.isnull().mean()*100
```

```python
import pandas as pd

def drop_columns(df):
    null_proportions = df.isnull().mean()
    cols_to_drop = null_proportions[null_proportions > 0.15].index
    df_dropped = df.drop(columns=cols_to_drop)
    
    return df_dropped
df.isnull().mean()*100
df_new=drop_columns(df)
```

```python
df_new
```
